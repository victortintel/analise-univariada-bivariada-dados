<p>Análise Univariada: Entender a distribuição e comportamento de uma única variável
<br>Análise Bivariada: Entender a relação entre duas variáveis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dados.csv')
df.head()
```

```python
# Verificando a dimensão do DataFrame
df.shape
```

### Análise Univariada
<p> Examinar e resumir os principais aspectos de cada variável

```python
# Medidas Estatísticas Básicas
print('Idade Mínima: ', df['idade'].min())
print('Idade Máxima: ', df['idade'].max())
print('Média da Idade: ', df['idade'].mean())
print('Mediana da Idade: ', df['idade'].median())
print('Desvio Padrão: ', df['idade'].std())
# A função mode retorna uma séries(indice + valor), por isso usamos o ILOC para pegar apenas o valor.
print('Moda: ', df['idade'].mode().iloc[0])
```

```python
# Verificando o boxplot da variavel idade
plt.figure(figsize=(8, 4))
sns.boxplot(df['idade'])
plt.title('Boxplot de Idade (Univariada)')
plt.ylabel('Idade')
plt.show()
```

```python
# Verificando o histograma da variavel idade
plt.figure(figsize=(8, 4))
sns.histplot(df['idade'], bins=30, kde=True)
plt.title('Distribuição de Idade (Univariada)')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

### Análise Bivariada
<p>Entendendo a relação da variável Salário e Idade

```python
# Verificando a relação entre a Variável Experência e Salário
plt.figure(figsize=(8, 4))
sns.regplot(data=df, x='experiencia', y='salario')
plt.title('Experiência vs Salário (Bivariada)')
plt.xlabel('Experiência (Anos)')
plt.ylabel('Salário R$')
plt.show()
```

- Cada ponto do gráfico acima é um registro do nosso conjunto de dados(uma observação). Essa linha que tem no gráfico se chama linha de regressão.

```python
# Verificando a relação entre a Variável Idade e Experiência
plt.figure(figsize=(8, 4))
sns.regplot(data=df, x='idade', y='experiencia')
plt.title('Idade vs Experiência (Bivariada)')
plt.xlabel('Idade (Anos)')
plt.ylabel('Experiência (Anos)')
plt.show()
```

```python
# Verificando a relação entre a Variável Idade e Salário
plt.figure(figsize=(8, 4))
sns.regplot(data=df, x='idade', y='salario')
plt.title('Idade vs Salário (Bivariada)')
plt.xlabel('Idade (Anos)')
plt.ylabel('Salário R$')
plt.show()
```

```python
df[['idade', 'experiencia', 'salario']].corr()
```

```python
# Calculando a correlação entre variáveis numéricas
correlacoes = df[['idade', 'experiencia', 'salario']].corr()

# Gerando o heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlacoes, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Mapa de Calor das Correlações')
plt.show()
```
